```python
import nbformat as nbf

# Read the .py script
with open('a2a_opto.py') as f:
    code = f.read()

# Create a new notebook object
nb = nbf.v4.new_notebook()

# Split the code by the #%% marker to identify cell breaks
code_blocks = code.split('#%%')

# Create cells for each code block
cells = [nbf.v4.new_code_cell(block) for block in code_blocks if block.strip()]

# Add cells to the notebook
nb['cells'] = cells

# Write the notebook to a file
with open('my_script.ipynb', 'w') as f:
    nbf.write(nb, f)
```
